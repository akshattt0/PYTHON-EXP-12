```python
import pandas as pd

data = {
    'Roll no' : [454,456,457,458,459],
    'Name' : ['Akshat', 'Max', 'Charles', 'Enzo', 'Michael'],
    'Age' : [23, 25, '-', 21, 29],
    'Marks' : [85, 90, 91, '-', 75],
    'Department' : ['CSE', 'Entc', 'IT', 'AIML', 'cse'],
    'Admission_Date': ['1/6/24', '3/6/24','5/6/24', '6/6/24','7-6-24']

}

df = pd.DataFrame(data)
df

```

```python
# Step 2: Replace '-'  with NaN to mark missing values

df.replace('-', np.nan, inplace=True)

# Convert columns to numeric where needed

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')
```

```python
# Step 3: Check missing values

print('\nMissing values: ')
print(df.isnull().sum())
```

```python
# Fill missing age with mean

df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing marks with median

df['Marks'].fillna(df['Marks'].median(), inplace=True)
df
```

```python
# Step 5: Fix inconsistent text formatting

df['Department'] = df['Department'].str.upper()
df
```

```python
# Step 6: Convert admission_date to standard datetime format

df['Admission_Date'] = pd.to_datetime(df['Admission_Date'], errors='coerce')
df
```

```python
# Step 7: Show cleaned data
print('\nCleaned data: ')
df
```

```python
# Step 8: Save cleaned dataset
df.to_csv('cleaned_student_data.csv', index=False)

print('\nCleaned CSV file saved as cleaned_student_data.csv')
```
